# Introduction to Transformer model

Till 2017-18, Recurrent neural network(RNN) were state of the art models for sequence modeling and transduction problems like language modeling and machine translation. To overcome this problem, transformer based models were introduced which instead of using recurrent network employs only attention mechanism and fully connected layers.

# Coding Task 1. SelfAttention

Self attention is computed as follows

$$
Attention(Q,K,V) = softmax(\frac{QK^{T}}{\sqrt{d_{k}}})V
$$

where K, Q, and V are all projections of your input

Q = Query matrix
K = Key matrix  
V = Value matrix  
$d_{k}$ = Square root of the dimension of the key or query vector.

Imagine you have two random matricies $K, Q \sim N(0, 1)$ of shape `[s, h]` and you compute $\operatorname{Softmax}(Q K^T)$.

**Q1:** How would softmax probabilities if we to increase $h$ to $2h$? Would they become more or less focused? Is is good for gradient flow?

> If you cannot show this mathematically, you can use pytorch and demonstrate it.

**Q2:** What happens if we divide attention scores by $\sqrt{h}$ before softmax?

**NOTE:** There are tips below.


### Your answers here

**A1:** Increasing h to 2h makes magnitudes and variance in attention scores larger, because what we perform to get softmax probabilities is basically a dot product. If we use high variance matrices K, Q for this dot product operation, the resultant matrix will be high variance matrix too. Therefore, we calculate softmax with larger values and variance; and it results in more focused softmax probabilities. It is bad for gradient flow because this situation results in extremely small gradients. Demonstration with pytorch:
```
>>> s, h = 3, 10
>>> a = torch.randn((s,h))
>>> b  = torch.randn((s,h))
>>> torch.softmax(torch.matmul(a,b.transpose(-1,-2)), dim=-1)
tensor([[0.5587, 0.0121, 0.4292],
        [0.0034, 0.9920, 0.0047],
        [0.4155, 0.0600, 0.5245]])

>>> a = torch.randn((s,2*h))
>>> b  = torch.randn((s,2*h))
>>> torch.softmax(torch.matmul(a,b.transpose(-1,-2)), dim=-1)
tensor([[7.5963e-06, 4.3286e-08, 9.9999e-01],
        [9.9993e-01, 5.5938e-05, 9.9217e-06],
        [9.8680e-01, 2.4163e-04, 1.2957e-02]])
```
You can see the case where s=3 and h=10. When we increase $h$ to $2h$, softmax probabilities get larger and one of the probabilities for each row is almost 1. Therefore, we can say that softmax probabilities become more focused. When all probabilities are almost 1 and 0, gradients are extremely small and it is not good for gradient flow.

**A2:** It will lower the variance and magnitude, so it fixes gradient flow by getting softmax function away from exteremely small gradient region.

# Coding task 2. Multi-head self-attention

Complete the **MultiheadSelfAttention(input_size, hidden_size, num_heads)** class in **layers.py** which accepts FloatTensor **[bs, seq_len, input_size]** and returns **FloatTensor[bs, seq_len, hidden_size]**. 'bs' stands for batch size and 'seq_len' is input sequence length. MultiHead attention is computed as below. 

**NOTE: Grades will be deducted if shapes of the tensors are not mentioned in your code**

Please read [Illustrated Transformer](https://jalammar.github.io/illustrated-transformer) for more details.

$$MultiHead(Q,K,V) = Concat(head_{1}, head_{2}, ..., head_{h})W^{O} $$ 

where$$ head_{i} = Attention(QW_{i}^{Q}, KW_{i}^{K}, VW_{i}^{V})$$

$W_{i}^{Q}$ = Weight matrix for $i^{th}$ head which will be multiplied to $i^{th}$ Query matrix  
$W_{i}^{K}$ = Weight matrix for $i^{th}$ head which will be multiplied to $i^{th}$ Key matrix  
$W_{i}^{V}$ = Weight matrix for $i^{th}$ head which will be multiplied to $i^{th}$ Value matrix  
$W^{O}$ = Weight  matrix which projects concatination of individual head attention into the final attention.

*   It should include **casual masking** - casual masking is applied to the scores (query -key product) to ensure that model does not pay attention to the future or current input tokens.


Note :

-  Remember to mix the heads before the output.
-  First implement multi head attention without casual masking. By default, the arguments for casual masking is set to False (`casual=False`).

### Inline questions

**Q3:** Do we scale the scores the same way for multi-head and single-head attention?

**A3:** For single-head, we scale sqrt of hidden size. For multi-head attention, we divide hidden size to num_heads and scale it with sqrt of head_size.

**Q4:** How many nonlinear operations do we have in self-attention? What about multi-head attention? Try to prove it.

**A4:** For self-attention, only 1 non-linear function (softmax) is performed for each word in each sequence. Therefore, (batch_size x seq_len) nonlinear operations are performed in self-attention. For multi-head attention, again only 1 non linear function is performed. But this time, it is performed for each word in each sequence and for each head. Therefore, (batch_size x num_heads x seq_len) nonlinear operations are performed in multi-head attention.

Using information from future words is cheating for a language model whose task is to predict the next word. Therefore, the situation we observe for the sequence 2 is something it shouldn't happen in a language model. When causal masking is on, we can look at the second sequence. It can be observed that attention scores for future words are all 0, which means cheating by looking future words is not allowed. Most of the words in the sequence yield high attention scores for the 2 which is allowed for a language model. word. It basically prevent information leaking from future words in the sequence.
